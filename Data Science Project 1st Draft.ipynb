### Hypothesis Testing

Null Hypothesis (H0): The model with three coefficients i.e. temperature, pressure, and dew point temperature is not significantly better than model with a single feature say temperature on influencing pollution levels (PM2.5).

Alternative Hypothesis (H1): The complex model is significantly better than the simple model on influencing pollution levels (PM2.5).

```python
merged_df[['DEWP','TEMP','PRES','RAIN']].corr()
```

```python
import statsmodels.api as sm

features = ['TEMP', 'PRES', 'DEWP','RAIN']
target_variable = 'PM2.5'

X = merged_df[features]
y = merged_df[target_variable]

X = sm.add_constant(X)
# print(X)
complex_model = sm.OLS(y, X).fit()

print(complex_model.summary())
```

```python
def compare_complex_with_one_feature_model(complex_model,features):

    X = merged_df[features]
    y = merged_df[target_variable]

    X = sm.add_constant(X)
    # print(X)
    simple_model = sm.OLS(y, X).fit()

    f_test = complex_model.compare_f_test(simple_model)

    f_statistic = f_test[0]
    p_value = f_test[1]

    alpha = 0.01

    # Determine whether to reject the null hypothesis
    if p_value < alpha:
        print(f"Reject the null hypothesis: The complex model with all features is better than the simple model with features : {features}")
    else:
        print("Fail to reject the null hypothesis: None of the predictors are significant.")

    print(f"F-statistic: {f_statistic}")
    print(f"P-value: {p_value}\n\n")
```

```python
from itertools import chain, combinations

def generate_all_proper_subsets_of_features(features):
  
  all_feature_combinations = list(chain.from_iterable(combinations(features, r) for r in range(len(features)+1)))
  all_feature_combinations = all_feature_combinations[1:-1]
  all_feature_combinations = [list(combination) for combination in all_feature_combinations]

  return all_feature_combinations
```

```python
all_feature_combinations = generate_all_proper_subsets_of_features(features)
```

```python
features = ['TEMP', 'PRES', 'DEWP','RAIN']

for feature_combination in all_feature_combinations:
    compare_complex_with_one_feature_model(complex_model,feature_combination)
```
